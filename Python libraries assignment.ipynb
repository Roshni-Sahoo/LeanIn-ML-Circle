```python
import numpy as np
arr=np.zeros(10)
print("Array of 10 0s=")
print(arr)
arr=np.ones(10)
print("Array of 10 1s=")
print(arr)
arr=np.ones(10)*5
print("Array of 10 5s")
print(arr)
arr=np.arange(30,70)
print("Array from 30 to 70")
print(arr)
```
