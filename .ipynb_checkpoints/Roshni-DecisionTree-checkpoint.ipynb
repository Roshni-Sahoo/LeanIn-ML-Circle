```python
# Load Libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
f=open('diabetes.csv','w')
f.close()
```

```python
##Data:https://www.kaggle.com/uciml/pima-indians-diabetes-database


# load dataset
pima = pd.read_csv("C:\\Users\\ROSHNI SAHOO\\Desktop\\leanin\\diabetes.csv")
```

```python
pima.head()
```

```python
pima.describe()
```

```python
#split dataset in features and target variable
#we keep all the attributes on one side and keep all the outcomes on a separate side.

X = pima.iloc[:, :-1].values
y = pima.iloc[:, 8].values
```

```python
#split dataset in features and target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
#random state=1 so that we get the same sequence of data when we open notebook.
```

```python
# Create Decision Tree classifier object
clf = DecisionTreeClassifier()
# Train Decision Tree Classifier
clf = clf.fit(X_train,y_train)
#Predict the response for test dataset
y_pred = clf.predict(X_test)
#we are inputting the x test values and we get the y predicted values.
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
#you are calculating it against what value in the test data and what value the model predicted.
```
